#  Python Variables

# A Variable is a name given to memory location in a program

Python variable name rules:-

Allowed:-
       
Letters: a–z, A–Z
       
-Numbers (but not at the start)
       
-Underscore _

Not allowed:-
       
-Special symbols:- @ # $ % & * - + etc.
       
-Spaces
       
-Starting with a number
