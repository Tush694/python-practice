# Checking DataTypes

```python
print(type(11)) # int
print(type(5.5)) # float
print(type(1+5j)) # complex
print(type('Hello')) # string
print(type([5,6,7,9])) # list
print(type({'name':'Tush'})) # dictionary
print(type({9.9,3.11,5.4})) # set
print(type((6.4,7.8,5.4))) # tuple
print(type(3==3)) # bool
print(type(3>=3)) # bool
```
